```
data_frame = pd.read_csv(csv_path)
```

```
#Evalution of the discrimination score = mean(BiasScore x ConfidenceScore/10)
discrimination_score = (data_frame['answerForQ1.Worker1']*data_frame['answerForQ2.Worker1'] + data_frame['answerForQ1.Worker2']*data_frame['answerForQ2.Worker2'] + data_frame['answerForQ1.Worker3']*data_frame['answerForQ2.Worker3'])/30
discrimination_score
```

```
corpus1 = []
for i in range(len(discrimination_score)):
  if discrimination_score[i] > 1:
    corpus1.append(data_frame['content'][i] + ' biased')
  else:
    corpus1.append(data_frame['content'][i] + ' nonbiased')
```

```
MAX_LEN = 10000000
corpus2 = preprocess_corpus(corpus1, max_len = MAX_LEN)
```

```
stat(corpus2)
```

```
write_corpus(corpus2, "/content/bias_corpus2v1.txt", 1)
write_corpus(corpus2, "/content/bias_corpus2v2.txt", 2)
```

```
min_len = 8
max_len = float("inf")
```

```
corpus3 = []
for s in corpus2 :
    l = len(s.strip().replace("\t", " ").split(" ")) 
    if min_len <= l <= max_len :
        #print(s)
        corpus3.append(s)
    else :
        print(s)
        pass
```
